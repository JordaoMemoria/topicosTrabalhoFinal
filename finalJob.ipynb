```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (8, 8)
mpl.rcParams['font.size'] = 14
mpl.rcParams['legend.fontsize'] = 14
```

## Lendo dataset

```python
df = pd.read_csv('houses_to_rent_v2.csv')
df
```

## Tornando o atributo city em 5 atributos categóricos binários

```python
df.city.unique()
```

```python
# Tornando o atributo city em 5 atributos categóricos binários

df['São Paulo'] = np.where(df['city']== 'São Paulo', 1, 0)
df['Porto Alegre'] = np.where(df['city']== 'Porto Alegre', 1, 0)
df['Rio de Janeiro'] = np.where(df['city']== 'Rio de Janeiro', 1, 0)
df['Campinas'] = np.where(df['city']== 'Campinas', 1, 0)
df['Belo Horizonte'] = np.where(df['city']== 'Belo Horizonte', 1, 0)

df
```

## Criando uma coluna nova y sendo ela a soma do condomínio + aluguel. Removendo colunas desnecessárias.

```python
df.columns
```

```python
# 
df['y con + alu'] = df['hoa (R$)'] + df['rent amount (R$)'] 
df = df.drop(columns=['city','hoa (R$)', 'rent amount (R$)', 'property tax (R$)', 'fire insurance (R$)', 'total (R$)'])
df['isHouse'] = df['floor'].apply(lambda x: 1 if x == '-' else 0)
df
```

## Colocando números em atributos categóricos

```python
df.loc[(df.animal == 'acept'),'animal'] = 1
df.loc[(df.animal == 'not acept'),'animal'] = 0

df.loc[(df.furniture == 'furnished'),'furniture'] = 1
df.loc[(df.furniture == 'not furnished'),'furniture'] = 0

# df.loc[(df.isHouse == '-'),'isHouse'] = 1
# df.loc[(df.isHouse != '-'),'isHouse'] = 0

df.loc[(df.floor == '-'),'floor'] = 1


df
```

## Modificando ordem das colunas apenas por estética subjetiva

```python
cols = df.columns.tolist()
cols
```

```python
cols = ['São Paulo', 'Porto Alegre', 'Rio de Janeiro', 'Campinas', 'Belo Horizonte',
     'area', 'rooms', 'bathroom', 'parking spaces', 'floor', 'isHouse', 'animal', 'furniture',
 'y con + alu']
df = df[cols]
df.describe()
```
